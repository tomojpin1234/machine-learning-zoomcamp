```python
# Q7. Sum of weights
```

```python
df.head()
```

```python
df_inn = df[df.Brand == "Innjoo"]
x = df_inn[["RAM", "Storage", "Screen"]].to_numpy()
```

```python
x
```

```python
XTX = x.T.dot(x)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
XTX_inv.dot(XTX)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100])
y
```

```python
w = XTX_inv.dot(x.T).dot(y)
w.sum()
```
