Create the correlation matrix for the numerical features of your dataset. In a correlation matrix, you compute the correlation coefficient between every pair of features.

## Question 5

- Let's find the least useful feature using the feature elimination technique.
- Train a model with all these features (using the same parameters as in Q4).
- Now exclude each feature from this set and train a model without it. Record the accuracy for each model.
- For each feature, calculate the difference between the original accuracy and the accuracy without the feature.


```python
from sklearn.metrics import accuracy_score

all_features = categorical+numerical

dv = DictVectorizer(sparse=False)
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)

train_dicts = df_train[all_features].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)
X_train

val_dicts = df_val[all_features].to_dict(orient='records')
X_val = dv.transform(val_dicts)
X_val

model.fit(X_train, y_train)

y_pred_elim = model.predict(X_val)
original_accuracy = accuracy_score(y_val, y_pred_elim)
print("Original accuracy ", original_accuracy)

answer_features = [
    "age",
    "balance",
    "marital",
    "previous",
]

accuracies = {}

for feature in answer_features:    
    feature_columns = [col for col in dv.get_feature_names_out() if col.startswith(feature)]    
    X_train_elim = X_train.copy()
    X_val_elim = X_val.copy()
    for col in feature_columns:        
        feature_index = np.where(dv.get_feature_names_out() == col)[0][0]        
        X_train_elim = np.delete(X_train_elim, feature_index, axis=1)
        X_val_elim = np.delete(X_val_elim, feature_index, axis=1)

    model_elim = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model_elim.fit(X_train_elim, y_train)
    
    y_pred_elim = model_elim.predict(X_val_elim)
    no_feature_accuracy = accuracy_score(y_val, y_pred_elim)
    accuracies[feature] = no_feature_accuracy    

accuracy_diffs_abs = {}
for feature, diff in accuracies.items():
    print(f"{feature}: {original_accuracy} - {diff} == {np.abs(original_accuracy - diff)}")
    accuracy_diffs_abs[feature] = np.abs(original_accuracy - diff)
```

The answer is `age`


Question 6

- Now let's train a regularized logistic regression.


```python
c = [0.01, 0.1, 1, 10, 100]
def train_model_with_regularization(c_value):
    dv = DictVectorizer(sparse=False)

    final_features = categorical+numerical

    train_dicts = df_train[final_features].to_dict(orient='records')
    X_train = dv.fit_transform(train_dicts)
    X_train

    val_dicts = df_val[final_features].to_dict(orient='records')
    X_val = dv.transform(val_dicts)
    X_val

    model = LogisticRegression(solver='liblinear', C=c_value, max_iter=1000, random_state=42)    
    model.fit(X_train, y_train)

    soft_prediction = model.predict_proba(X_val)
    y_pred = soft_prediction[:, 1]
    sub_predict = (y_pred >= 0.5).astype(int)
    # Accuracy
    return round((sub_predict == y_val).mean(), 3)

for c_value in c:
    accuracy = train_model_with_regularization(c_value)  
    print(f"c = {c_value}: accuracy = {accuracy}")
```

The answer is `0.1`
